```python
html = urlopen('http://gosi.go.kr/')
bsObject = BeautifulSoup(html, "html.parser")
```

```python
bsObject
```

```python
bsObject.body
```

```python
for n in range(1,13):
    div = bsObject.body.find('div',{'id':'box'+str(n)})
    종류 = div.caption.get_text().split()[0]
    원서접수 = div.select('td')[0].get_text()
    면접시험 = div.select('td')[-2].get_text()
    최종합격자발표 = div.select('td')[-1].get_text()
    print(종류, 원서접수, 면접시험, 최종합격자발표)
```

```python
# Create a file to write to, add headers row
with open('crawling_gosi.go.kr.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['종류', '원서접수', '면접시험', '최종합격자발표'])
        
    for n in range(1,13):
        div = bsObject.body.find('div',{'id':'box'+str(n)})
        종류 = div.caption.get_text().split()[0]
        원서접수 = div.select('td')[0].get_text()
        면접시험 = div.select('td')[-2].get_text()
        최종합격자발표 = div.select('td')[-1].get_text()
        writer.writerow([종류, 원서접수, 면접시험, 최종합격자발표])
```
